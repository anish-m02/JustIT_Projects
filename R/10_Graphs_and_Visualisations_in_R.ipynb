###Plot()


The `plot()` function is used to draw points (markers) in a diagram.

The function takes parameters for specifying points in the diagram.

Parameter 1 specifies points on the **x-axis**.

Parameter 2 specifies points on the **y-axis**.

At its simplest, you can use the `plot()` function to plot two numbers against each other:

### Example

```r
# Create a simple scatter plot with a single point at coordinates (1, 3)
plot(1, 3)
```

1. **Creating a Scatter Plot:**
   - The `plot()` function is used to create a scatter plot.
   - In this case, we provide the x-coordinate as `1` and the y-coordinate as `3`.
   - The result will be a scatter plot with a single point at the specified position.

2. **Result:**
   - The plot will display a single point at coordinates (1, 3).

### Size `cex=_number_`

Use `cex=_number_` to change the size of the points (`1` is default, while `0.5` means 50% smaller, and `2` means 100% larger).

The **`cex`** parameter in R stands for **"character expansion"**. It is used to control the size of various graphical elements, such as symbols (points) and text, relative to the default size. Specifically:

1. **Symbols (Points) Size:**
   - The `cex` parameter changes the size of symbols (points) in plots.
   - By default, the value of `cex` is 1.
   - A value greater than 1 (e.g., 2) will increase the size of symbols.
   - A value less than 1 (e.g., 0.5) will decrease the size of symbols.

2. **Text Size:**
   - The `cex` parameter also affects the size of text elements in plots.
   - For example:
     - `cex.main`: Changes the size of the plot title.
     - `cex.lab`: Changes the size of x-axis and y-axis labels.
     - `cex.axis`: Changes the size of axis tick mark annotations.

Feel free to adjust the `cex` value to create plots with the exact sizes you desire! ðð


```r
# Create a scatter plot with x-coordinates ranging from 1 to 10
# The 'cex' parameter is set to 2, which controls the size of the points
plot(1:10, cex=2)
```

1. **Creating a Scatter Plot:**
   - The `plot()` function is used to create a scatter plot.
   - We provide a single argument: `1:10`.
   - The `1:10` expression generates a sequence of integers from 1 to 10.

2. **Customizing the Plot (Point Size):**
   - The `cex` parameter specifies the size of the points in the scatter plot.
   - A value of 2 means that the points will be twice as large as the default size.

3. **Result (Expected Plot):**
   - The plot will display points along the x-axis from 1 to 10.
   - All points will be larger than the default size due to the `cex=2` setting.

The resulting scatter plot would show larger points at positions (1, 1), (2, 2), ..., (10, 10). ðð Â¹.

Line Graphs
-----------

A line graph has a line that connects all the points in a diagram.

To create a line, use the `plot()` function and add the `type` parameter with a value of `"l"`:



```r
# Example R code: Plotting a line graph
# We'll create a simple line plot using the plot() function

# Generate a sequence of numbers from 1 to 10
x_values <- 1:10

# Plot the sequence against itself (y = x)
# The 'type' argument specifies that we want a line plot ('l')
plot(x_values, type = "l")
```

Here's what each part of the code does:

1. **Creating a Sequence (`x_values`)**:
   - We create a numeric vector called `x_values` containing the numbers from 1 to 10 using the `1:10` notation.
   - This vector represents the x-axis values for our plot.

2. **Plotting the Line (`plot(x_values, type = "l")`)**:
   - The `plot()` function is a versatile tool for creating various types of graphs in R.
   - In this case, we use it to create a **line plot**.
   - The `x_values` serve as the x-coordinates, and the corresponding y-coordinates are assumed to be the same (i.e., y = x).
   - The `type = "l"` argument specifies that we want to connect the points with a line.
   - The resulting plot will display a straight line connecting the points (1,1), (2,2), ..., (10,10).

Line Width
----------

To change the width of the line, use the `lwd` parameter (`1` is default, while `0.5` means 50% smaller, and `2` means 100% larger):


```r
# Example R code: Plotting a line graph
# We'll create a simple line plot using the plot() function

# Generate a sequence of numbers from 1 to 10
x_values <- 1:10

# Plot the sequence against itself (y = x)
# The 'type' argument specifies that we want a line plot ('l')
# The 'lwd' argument sets the line width to 2 (thicker line)
plot(x_values, type = "l", lwd = 2)
```

Here's what each part of the code does:

1. **Creating a Sequence (`x_values`)**:
   - We create a numeric vector called `x_values` containing the numbers from 1 to 10 using the `1:10` notation.

Pie Charts

A pie chart is a circular graphical view of data.


Here's what each part of the code does:

   - We define a numeric vector called `x`.
   - The vector contains the values `(10, 20, 30, 40)`.

2. **Plotting the Pie Chart (`pie(x)`)**:
   - The `pie()` function creates a pie chart.
   - We pass the vector `x` as input to the function.
   - The resulting pie chart displays slices corresponding to the proportions `(10%, 20%, 30%, 40%)`.

The pie chart visually represents the distribution of values in the vector `x`. Each slice represents a portion of the whole, and the size of each slice corresponds to the value in the vector. ð¥§ð

````r
# Example R code: Creating a pie chart
# We'll use the pie() function to create a simple pie chart

# Create a vector of values representing the sizes of pie slices
x <- c(10, 20, 30, 40)

# Display the pie chart using the pie() function
pie(x)
````

#### Example Explained

As you can see the pie chart draws one pie for each value in the vector (in this case 10, 20, 30, 40).

By default, the plotting of the first pie starts from the x-axis and move **counterclockwise**.

**Note:** The size of each pie is determined by comparing the value with all the other values, by using this formula:

The value divided by the sum of all values: `x/sum(x)`

Bar Charts
----------

A bar chart uses rectangular bars to visualize data. Bar charts can be displayed horizontally or vertically. The height or length of the bars are proportional to the values they represent.

Use the `barplot()` function to draw a vertical bar chart:

### Example


```r
# Example R code: Creating a bar plot

# Create a vector of x-axis labels
x <- c("A", "B", "C", "D")

# Create a vector of y-axis values
y <- c(2, 4, 6, 8)

# Display the bar plot using the barplot() function
barplot(y, names.arg = x)
```

Here's what each part of the code does:

1. **Creating a Vector of X-Axis Labels (`x`)**:
   - We define a character vector called `x`.
   - The vector contains the labels `"A"`, `"B"`, `"C"`, and `"D"` corresponding to the x-axis categories.

2. **Creating a Vector of Y-Axis Values (`y`)**:
   - We define a numeric vector called `y`.
   - The vector contains the values `(2, 4, 6, 8)` corresponding to the heights of the bars.

3. **Plotting the Bar Chart (`barplot(y, names.arg = x)`)**:
   - The `barplot()` function creates a bar plot.
   - We pass the vector `y` as input to the function.
   - The `names.arg` argument specifies the labels for the x-axis categories.

The resulting bar plot visually represents the data with bars corresponding to each category (A, B, C, D) and their respective heights. ðð

Developed by **Hadley Wickham** in 2005, ggplot2 implements **Leland Wilkinson's Grammar of Graphics**, which provides a structured approach to creating graphs by breaking them down into semantic components such as scales and layersÂ¹Â².

Here are the key points about ggplot2:

1. **Declarative Approach**: With ggplot2, you provide the data, specify how to map variables to aesthetics (such as color or size), choose graphical primitives (like points or bars), and it takes care of the details. This declarative approach allows you to create complex visualizations with ease.

2. **Abstraction Layers**: ggplot2 allows you to add, remove, or alter components in a plot at a high level of abstraction. You start with `ggplot()`, supply a dataset and aesthetic mappings using `aes()`, and then add layers (e.g., `geom_point()` for scatter plots or `geom_histogram()` for histograms), scales, faceting specifications, and coordinate systems.

###Key Terminologies in ggplot2

In ggplot2, `aes` stands for **aesthetic mappings**. It's a crucial function that defines how variables in your data are mapped to visual properties (aesthetics) within your plot.

Here's a breakdown of how `aes` works:

**Syntax:**

```R
ggplot(data, aes(x = variable1, y = variable2, ...)) + geom_function()
```

- `data`: Your data frame containing relevant variables.
- `aes()`: The aesthetic mapping function.
- `x = variable1`, `y = variable2`: These define which variables are mapped to the x and y axes, respectively. You can map additional aesthetics like `color`, `fill`, `size`, `alpha`, etc., within the same `aes()` function.
- `geom_function()`: This specifies the geometric object (e.g., `geom_point()`, `geom_bar()`, `geom_line()`) you want to use to represent the data points.

**Essentially, `aes` acts as a bridge between your data and the visual elements of your plot.** It tells ggplot2 how to translate numerical values from your variables into visual properties like color, position, size, transparency, and so on.

**Example:**

```R
ggplot(iris, aes(x = Sepal.Length, y = Petal.Length, color = Species)) + geom_point()
```

Here, `aes` maps the "Sepal.Length" variable to the x-axis, "Petal.Length" to the y-axis, and "Species" to the color aesthetic, resulting in a scatter plot where points are colored based on their species.

Remember, you can experiment with different aesthetic mappings to create diverse and insightful visualizations. For deeper understanding, refer to the ggplot2 documentation and explore various examples online!
