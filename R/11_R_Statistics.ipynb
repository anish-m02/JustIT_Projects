Data Set
--------

A data set is a collection of data, often presented in a table.

There is a popular built-in data set in R called "**mtcars**" (Motor Trend Car Road Tests), which is retrieved from the 1974 Motor Trend US Magazine.

The `mtcars` dataset is a built-in dataset in R that contains measurements on 11 different attributes for 32 different cars. Here's what each of the columns in the dataset represents:

1. `mpg`: Miles/(US) gallon
2. `cyl`: Number of cylinders
3. `disp`: Displacement (cu.in.)
4. `hp`: Gross horsepower
5. `drat`: Rear axle ratio
6. `wt`: Weight (1000 lbs)
7. `qsec`: 1/4 mile time
8. `vs`: V/S
9. `am`: Transmission (0 = automatic, 1 = manual)
10. `gear`: Number of forward gears
11. `carb`: Number of carburetors

You can print the `mtcars` dataset in R by simply typing `mtcars` in the console. This will display the entire dataset in the console. ðŸš—ðŸ“Š

#R Quantiles

### Main Concepts

Let's clear up the main concepts first by breaking down the differences between **quantiles**, **percentiles**, **percentages**, and **quartiles**:

1. **Quantiles**:
   - Quantiles divide a dataset into equal parts.
   - They can range from any value to any other value.
   - Examples of quantiles include quartiles (4-quantiles), quintiles (5-quantiles), deciles (10-quantiles), and percentiles (100-quantiles).
   - Quantiles are typically expressed as decimal values and range from 0 to 1.

2. **Percentiles**:
   - Percentiles are a specific type of quantile.
   - They divide a dataset into 100 equal parts.
   - Percentiles range from 0 to 100.
   - For example, the 90th percentile represents the value below which 90% of the data falls.

3. **Percentages**:
   - Percentages are not directly related to quantiles.
   - They express a proportion or relative part of a whole.
   - Percentages are always expressed as a percentage (e.g., 50%, 75%, etc.).
   - For example, if 75% of students passed an exam, it means 75 out of every 100 students passed.

4. **Quartiles**:
   - Quartiles are also a specific type of quantile.
   - They divide a dataset into four equal parts.
   - Quartiles range from 0 to 4.
   - The 1st quartile (25th percentile) separates the bottom 25% of values from the top 75%.
   - The 2nd quartile (50th percentile) is the median, dividing the data into equal halves.
   - The 3rd quartile (75th percentile) separates the bottom 75% of values from the top 25%.

In summary:
- **Quantiles** are a general concept that includes percentiles and quartiles.
- **Percentiles** divide data into 100 equal parts (e.g., 90th percentile).
- **Quartiles** divide data into 4 equal parts (e.g., 1st quartile, 2nd quartile, 3rd quartile).
